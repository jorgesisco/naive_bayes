# Naive Bayes
Load the `mnist` dataset. Split it into training and test sets. Train and test a naive bayes model using scikit-learn. Check the documentation to identify the most important hyperparameters, attributes, and methods of the model. Use them in practice.

## Importing dependencies

```python
import pandas as pd #load data set
from sklearn.naive_bayes import GaussianNB  # Bayes!!
import sklearn.model_selection #split the data!!
import sklearn.metrics # to calculate accuracy!!!!
```

## Import the data base

```python
df = pd.read_csv('/mnist.csv')
df = df.set_index('id')
df
```

## Spliting Dataset train and test datasets

```python
x = df.drop(['class'], axis=1)
y =df['class']

x_train, x_test, y_train, y_test, = sklearn.model_selection.train_test_split(x,y)

print('df', df.shape)
print('x_train', x_train.shape)
print('x_test', x_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

## Training model

```python
clf = GaussianNB()
clf.fit(x_train, y_train)
```

## Testing the trained model

```python
y_predicted = clf.predict(x_test)
y_predicted
```

## testing accuracy 

```python
accuracy = sklearn.metrics.accuracy_score(y_test, y_predicted)
accuracy
```
